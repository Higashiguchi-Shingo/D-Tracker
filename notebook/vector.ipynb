```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def rds(W, A, D):
    return A[0] * W[0] + D[0] * (W[1] - W[0]), A[1] * W[1] + D[1] * (W[0] - W[1])

def vecfield(A, D, trajectory):
    W = list(np.meshgrid(np.linspace(0,0.12, 10), np.linspace(0,0.04, 10)))
    U, V = rds(W, A, D)

    plt.figure(figsize=(8, 8))
    plt.quiver(W[0], W[1], U, V, color='black')
    plt.plot(trajectory[:, 0], trajectory[:, 1], 'r', label='Trajectory', linewidth=5)
    plt.xlabel('w0')
    plt.ylabel('w1')
    plt.title('Vector Field')
    plt.grid()
    plt.show()

```

```python
path = "../result/device/lc=104/stl=26/rankupdate=1/ablation_s=0/ablation_d=0/outlier=1/model"
A = np.load(path + "/t=44-/A.npy")
D = np.load(path + "/t=44-/D.npy")
Wcore = np.load(path + "/t=44-/Wcore.npy")
D = [D[0][0][1], D[0][1][0]]
Wcore = Wcore[:,0,:]
```

```python
vecfield(A[0], D, Wcore)
```
