```python
dataset = "device"
dir = "../result/" + dataset + "/lc=104/stl=26/rankupdate=1/ablation_s=0/ablation_d=0/outlier=1/forecast"

X = np.array([])
X_hat = np.array([])

def concat(X, X_dt):
    if len(X)==0:
        return X_dt
    else:
        return np.concatenate([X, X_dt], axis=0)
```

```python
ts = 312
te = 636
dt = 4
l = 13

for t in range(ts, te, dt):
    path = dir + "/t=" + str(t)
    Xf = np.load(path + "/Xf.npy")[l-dt:l]
    Xf_hat = np.load(path + "/Xf_hat.npy")[l-dt:l]

    X = concat(X, Xf)
    X_hat = concat(X_hat, Xf_hat)
```
